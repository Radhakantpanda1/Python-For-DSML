# Return Statement

It is used to exit a function and go to the place from where it is called

```python
def getSum(arr):
    sum=0
    for i in range(len(arr)):
        sum+=arr[i]
    return sum
sum=getSum([1,2,3,4,5,6,7,8,9])
print(sum)
```

# enumerate()

```python
listA=['a','b','c','d','e']
for index,val in enumerate(listA):
    print(f'The index is {index} and the value is {val}')
```

```python
listA=['a','b','c','d','e']
for index,val in enumerate(listA,10):# we can give any statrting index of our choice
    print(f'The index is {index} and the value is {val}')
```

```python
def pos_num(num):
    if num>0:
        return num

number_list=range(-10,11)
print(list(number_list))
positive_number_list=list(filter(pos_num,number_list))
print(positive_number_list)
```

```python
listA=[1,2,3,4,5]
print(isinstance(listA,list))
```

```python
tplA=tuple([1,2,3,4,5])
print(isinstance(tplA,list))
```

```python
def make_square(n):
    return n**2
square_list=list(map(make_square,listA))
print(square_list)
```

```python
from functools import reduce
def multiplly(a,b):
    return a*b
ans=reduce(multiplly,listA)
print(ans)
```

```python
def sum_arr(x,y):
    return x+y
summ=reduce(sum_arr,listA)
print(summ)
```

# Function Arguments
