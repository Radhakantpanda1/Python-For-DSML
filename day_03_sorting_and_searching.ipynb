# Insertion Sort (stable algorithm , online algorithm)

## price is the key  rest all are the satellite data (because it moves along with the key)

## Code

```python
def insertionSort(arr):
    i=1
    while i<len(arr):
        j=i
        k=arr[i]
        while j>0 and arr[j-1]>k:
            arr[j]=arr[j-1]
            j-=1
        arr[j]=k
        i+=1
    return arr;
x=insertionSort([6,5,3,1,8,7,2,4])
print(x)
            
                
                

        
```
